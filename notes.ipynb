```python
import pandas as pd 
import numpy as np 
```

```python
def read_csv_to_dataframe(file_path):
    try:
        # Read the CSV file into a Pandas DataFrame
        dataframe = pd.read_csv(file_path)
        dataframe.bfill(inplace=True)
        return dataframe
    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return None
    except Exception as e:
        print(f"An error occurred: {str(e)}")
        return None
```

```python
# Load patients, inputevents, and vital signs dataframe from MIMIC-IV 

discharge_df = read_csv_to_dataframe("data/discharge.csv")
```

```python
def get_discharge_notes_by_subject_ids(subject_ids, df):
    # Use the isin method to filter the DataFrame based on subject_ids
    filtered_df = df.iloc[:100][df['subject_id'].isin(subject_ids)]

    # Extract the 'text' column values as a NumPy array
    text_values = filtered_df['text'].to_numpy()

    return text_values

# Example usage
subject_ids_to_search = [10000032]
resulting_texts = get_discharge_notes_by_subject_ids(subject_ids_to_search, discharge_df)

print(resulting_texts)
print(len(resulting_texts))
```

```python
def clean_and_save_text(output_file, text_values):
    # Join the text values into a single string
    text_combined = '\n'.join(text_values)

    # Remove '\n' characters
    cleaned_text = text_combined.replace('\n', ' ')

    # Save the cleaned text to a file
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(cleaned_text)

# Example usage
output_file_path = 'output.txt'
subject_ids_to_search = [10000032]
filtered_df = discharge_df[discharge_df['subject_id'].isin(subject_ids_to_search)]
text_values = filtered_df['text'].to_numpy()

clean_and_save_text(output_file_path, text_values)
```
